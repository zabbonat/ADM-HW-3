```python
#Save the title and paragraphs on 'prova' ['first three links of movies1.html']
#try with everything
title=[]
paragraphs=[]
for i in range(len(movies_1_link)):
    response= requests.get(movies_1_link[i])
    if response is not None:
        page=bs4.BeautifulSoup(response.text, 'lxml-xml')
        title.append(page.select("#firstHeading")[0].text)
        paragraphs.append(page.select("p"))
    print(i)

    


#tsv file
```
